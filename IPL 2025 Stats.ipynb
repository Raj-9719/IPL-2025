# IPL 2025 Players Performance Analysis

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
IPLdata_1 = pd.read_csv("Stats.csv")
IPLdata_1
```

# Who were the top 10 run-scorers in IPL 2025?

```python
top10 = IPLdata_1.sort_values(by='Total_runs_scored',ascending=False).head(10)
plt.figure(figsize=(8,4))
plt.bar(top10['Player_name'], top10['Total_runs_scored'], color= 'orange')
plt.title("Top 10 Most Runs Scored Batsman in  IPL 2025")
plt.xlabel("Player Name")
plt.ylabel("Total Runs")
plt.xticks(rotation = 45)
plt.show()
```

# Who were the top 10 wicket-takers in IPL 2025?

```python
top10 = IPLdata_1.sort_values(by='Wkts_taken',ascending=False).head(10)
plt.figure(figsize=(8,4))
plt.bar(top10['Player_name'], top10['Wkts_taken'], color='purple')
plt.title("Top 10 Most Wicket taken Bowler in  IPL 2025")
plt.xlabel("Player Name")
plt.ylabel("Total Wickets")
plt.xticks(rotation = 45)
plt.show()
```

#  Who had the top 10 best economy bowlers?

```python
bowlers = IPLdata_1[(IPLdata_1['Category'] == 'Bowler') & (IPLdata_1['Wkts_taken'] > 0)]

# Sort by Economy (ascending → best economy first)
top10 = bowlers.sort_values(by='Economy', ascending=True).head(10)

plt.figure(figsize=(8,4))
plt.bar(top10['Player_name'], top10['Economy'], color='royalblue')
plt.title("Top 10 Best Economical Bowlers in  IPL 2025")
plt.xlabel("Player Name")
plt.ylabel("Economy")
plt.xticks(rotation = 45)
plt.show()
```

# Who were the top 10 all-rounders in IPL 2025?

```python
all_rounders = IPLdata_1[IPLdata_1['Category'] == 'AR'].copy()

# Create a combined score formula
all_rounders["AllRounder_Score"] = (all_rounders["Total_runs_scored"] + (all_rounders["Wkts_taken"] * 20) + (all_rounders["Catch_taken"] * 5))

# Sort by Score in descending order
all_rounders_sorted = all_rounders.sort_values(by ="AllRounder_Score", ascending=False)

# Select only the needed columns
all_rounders_df = all_rounders_sorted[["Player_name","Team","Total_runs_scored","Highest_score","Wkts_taken","Catch_taken","Best_Bowling","AllRounder_Score"]].reset_index(drop=True)

# Top 10 all-Rounders
top_10_all_rounders = all_rounders_df.head(10)
print(top_10_all_rounders)
```

```python
plt.figure(figsize=(8,4))
plt.bar(top_10_all_rounders['Player_name'], top_10_all_rounders['AllRounder_Score'],color='magenta')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Player Name")
plt.ylabel("All Rounder Score")
plt.title("Top 10 All-Rounders(Score based) in IPL 2025")
plt.show()
```
