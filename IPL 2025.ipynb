# IPL 2025 Performance Analysis

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
IPLdata = pd.read_csv("Matches.csv")
IPLdata
```

# How many matches did each team win in IPL 2025?

```python
IPLdata['match_winner'].value_counts()
plt.figure(figsize=(10,6))
ax = sns.countplot( x='match_winner', data=IPLdata, hue='match_winner', palette="viridis", legend=False)
ax.set_title("Number of Matches Won by Each Team")
ax.set_xlabel("Team Name")
ax.set_ylabel("Number of Matches")
plt.xticks(rotation=45)

plt.show()
```

# Which teams won the toss most often?

```python
IPLdata['toss_winner'].value_counts()
plt.figure(figsize=(10,6))
ax=sns.countplot(x = 'toss_winner', data= IPLdata,hue='toss_winner', palette="crest", legend=False)
ax.set_title("Number of Times Toss Won by Each Team")
ax.set_xlabel("Team Name")
ax.set_ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# What decisions did teams make after winning the toss?

```python
IPLdata['toss_decision'].value_counts()
plt.figure(figsize=(10,6))
ax = sns.countplot( x = 'toss_decision', data=IPLdata,hue='toss_decision', palette="magma", legend=True)
ax.set_title("Decision Taken by Teams After Winning the Toss")
ax.set_xlabel("Decision")
ax.set_ylabel("Counts")
plt.show()
```

# How often did the toss-winning team also win the match?

```python
(IPLdata['toss_winner']==IPLdata['match_winner']).value_counts()
```

# How many matches were won by teams batting first?

```python
team_1_won=IPLdata[(IPLdata['first_ings_score']>IPLdata['second_ings_score'])&(IPLdata['toss_decision']=='Bat')]
team_1_won
```

# How many matches were won by teams fielding first?

```python
team_2_won=IPLdata[(IPLdata['first_ings_score']<IPLdata['second_ings_score'])&(IPLdata['toss_decision']=='Field')]
team_2_won
```

# Who won the most "Player of the Match" awards in IPL 2025 (Top 10)?

```python
list(IPLdata["player_of_the_match"].value_counts()[0:10].keys())
plt.figure(figsize=(8,4))
plt.bar(list(IPLdata['player_of_the_match'].value_counts() [0:10].keys()),IPLdata['player_of_the_match'].value_counts()[0:10])
plt.title("Top 10 Player of the Match in IPL 2025")
plt.xlabel("Player Name")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# Which team had the highest win percentage?

```python
matches_played = pd.concat([IPLdata['team1'],IPLdata['team2']]).value_counts()
wins= IPLdata['match_winner'].value_counts()
win_percentage = (wins/ matches_played * 100).round(2).sort_values(ascending=False).head(10)
df_win = win_percentage.reset_index()
df_win.columns = ['Team', 'Win_Percentage']
plt.figure
sns.barplot(data= df_win,x= 'Win_Percentage', y= 'Team',hue= 'Team', palette='crest', legend=False)
plt.title("Win Percentage by Team")
plt.xlabel("Win Percentage")
plt.ylabel("Team")
plt.show()
```

# Did winning the toss influence match results?

```python
toss_wins = (IPLdata['toss_winner'] == IPLdata['match_winner']).sum()
toss_losses = len(IPLdata)- toss_wins
plt.figure
plt.pie([toss_wins, toss_losses], labels=["Toss & Match Won", "Toss Won, Match Lost"],
        autopct= "%1.1f%%", colors= ["#4CAF50", "#FF5252"])
plt.title("Impact of Toss on Match Result")
plt.show()
```
